LAMPIRAN KODE NUMERIK PYTHON METODE PERCOBAAN PATTERN SEARCH

```python

import numpy as np

# ===============================
# Fungsi Rastrigin
# ===============================
def rastrigin(x):
    n = len(x)
    return np.sum(x**2 - 10*np.cos(2*np.pi*x) + 10)


# ===============================
# Pattern Search Algorithm
# ===============================
def pattern_search(
    f,
    x0,
    delta0=0.5,
    beta=0.5,
    max_iter=1000,
    tol=1e-6,
    lower_bound=-5,
    upper_bound=5
):
    x = x0.copy()
    delta = delta0
    n = len(x)

    best_value = f(x)

    for k in range(max_iter):
        improved = False

        # Exploratory move
        for i in range(n):
            for direction in [+1, -1]:
                x_new = x.copy()
                x_new[i] += direction * delta

                # Boundary handling
                x_new[i] = np.clip(x_new[i], lower_bound, upper_bound)

                f_new = f(x_new)

                if f_new < best_value:
                    x = x_new
                    best_value = f_new
                    improved = True

        # Update step size
        if not improved:
            delta *= beta

        # Stopping criterion
        if delta < tol:
            break

    return x, best_value, k + 1


# ===============================
# Parameter Umum
# ===============================
DIMENSION = 30
MAX_ITER = 1000
TOL = 1e-6


# ===============================
# Percobaan 1: Variasi Δ0
# ===============================
print("=== Percobaan 1: Variasi Delta Awal ===")

delta_values = [0.1, 0.5, 1.0]
beta_fixed = 0.5

for delta0 in delta_values:
    x0 = np.random.uniform(-5, 5, DIMENSION)
    x_best, f_best, iters = pattern_search(
        rastrigin, x0, delta0, beta_fixed, MAX_ITER, TOL
    )
    print(f"Delta0 = {delta0:>4} | f_min = {f_best:.6f} | iter = {iters}")


# ===============================
# Percobaan 2: Variasi β
# ===============================
print("\n=== Percobaan 2: Variasi Beta ===")

beta_values = [0.3, 0.5, 0.8]
delta_fixed = 0.5

for beta in beta_values:
    x0 = np.random.uniform(-5, 5, DIMENSION)
    x_best, f_best, iters = pattern_search(
        rastrigin, x0, delta_fixed, beta, MAX_ITER, TOL
    )
    print(f"Beta = {beta:>4} | f_min = {f_best:.6f} | iter = {iters}")


# ===============================
# Percobaan 3: Pengaruh Solusi Awal
# ===============================
print("\n=== Percobaan 3: Pengaruh Solusi Awal ===")

delta0 = 0.5
beta = 0.5

for run in range(1, 6):
    x0 = np.random.uniform(-5, 5, DIMENSION)
    x_best, f_best, iters = pattern_search(
        rastrigin, x0, delta0, beta, MAX_ITER, TOL
    )
    print(f"Run {run} | f_min = {f_best:.6f} | iter = {iters}")


# ===============================
# Percobaan 4: Kombinasi Parameter Terbaik
# ===============================
print("\n=== Percobaan 4: Kombinasi Parameter Terbaik ===")

best_delta = 0.5
best_beta = 0.5

for run in range(1, 6):
    x0 = np.random.uniform(-5, 5, DIMENSION)
    x_best, f_best, iters = pattern_search(
        rastrigin, x0, best_delta, best_beta, MAX_ITER, TOL
    )
    print(f"Final Run {run} | f_min = {f_best:.6f} | iter = {iters}")
```
